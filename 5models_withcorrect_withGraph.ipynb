```python
import gradio as gr
import tempfile
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample synthetic data for Kerala districts and baseline flood probability (you can customize or load from CSV)
district_data = pd.DataFrame({
    'District': ['Thiruvananthapuram', 'Kollam', 'Pathanamthitta', 'Alappuzha', 'Kottayam', 'Idukki', 'Ernakulam',
                 'Thrissur', 'Palakkad', 'Malappuram', 'Kozhikode', 'Wayanad', 'Kannur', 'Kasaragod'],
    'Base_Flood_Probability': [45, 50, 60, 55, 58, 65, 62, 55, 59, 48, 52, 47, 49, 46]
})

def get_weather_data(latitude=10.8505, longitude=76.2711):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&current_weather=true"
    response = requests.get(url)
    if response.status_code != 200:
        return "Weather data not available"
    data = response.json()
    temp = data["current_weather"]["temperature"]
    windspeed = data["current_weather"]["windspeed"]
    return f"Kerala: {temp}C, Wind Speed: {windspeed} km/h"

# Load your model (make sure to load it)
import joblib


def plot_severity(probability):
    # Update severity for each district by amplifying with prediction probability
    district_data['Adjusted_Probability'] = district_data['Base_Flood_Probability'] * (probability / 100)

    def categorize(p):
        if p > 60:
            return 'High Severity'
        elif p > 40:
            return 'Moderate Severity'
        else:
            return 'Low Severity'

    district_data['Severity'] = district_data['Adjusted_Probability'].apply(categorize)

    # Plot
    plt.figure(figsize=(12, 8))
    sns.barplot(
        x='Adjusted_Probability', y='District', hue='Severity',
        data=district_data.sort_values('Adjusted_Probability'),
        dodge=False, palette={'Low Severity': 'green', 'Moderate Severity': 'orange', 'High Severity': 'red'}
    )
    plt.xlabel('Adjusted Flood Probability (%)')
    plt.title('District-wise Severity Prediction')
    plt.legend(title='Severity Level')

    tmpfile = tempfile.NamedTemporaryFile(suffix=".png", delete=False)
    plt.savefig(tmpfile.name)
    plt.close()
    return tmpfile.name

def predict_flood_and_visualize(year, jan, feb, mar, apr, may, jun, jul, aug, sep, oct_, nov, dec):
    user_values = [[year, jan, feb, mar, apr, may, jun, jul, aug, sep, oct_, nov, dec]]
    prediction = lr.predict(user_values)
    probability = lr.predict_proba(user_values)[0][1] * 100

    if prediction[0] == 1:
        result_text = f" Flood is likely to occur.\nEstimated probability: {probability:.2f}%"
    else:
        result_text = f" No flood expected.\nEstimated probability of flood: {probability:.2f}%"

    # CSV output
    df = pd.DataFrame(user_values, columns=['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'])
    df['Flood_Predicted'] = 'Yes' if prediction[0] == 1 else 'No'
    df['Probability'] = f"{probability:.2f}%"

    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix=".csv")
    df.to_csv(temp_file.name, index=False)

    # Generate severity graph image
    severity_graph = plot_severity(probability)

    return result_text, temp_file.name, severity_graph

# CSS customization
custom_css = """
body {
    background-color: #1397f0;
}
.gradio-container {
    background-color: #f0f9ff !important;
}
button {
    background-color: #0c95f0 !important;
    color: white !important;
}
.weather-box {
    background-color: #ffffffcc;
    padding: 8px 15px;
    border-radius: 10px;
    font-weight: bold;
    color: #333;
    margin-left: auto;
}
"""

current_weather = get_weather_data()

with gr.Blocks(css=custom_css) as demo:
    with gr.Row():
        gr.Markdown("##  Flood Prediction System ")
        gr.HTML(f"<div class='weather-box'> {current_weather}</div>")

    year = gr.Number(label=" Year")

    with gr.Row():
        jan = gr.Number(label=" JANUARY")
        feb = gr.Number(label=" FEBRUARY")
        mar = gr.Number(label=" MARCH")
        apr = gr.Number(label=" APRIL")

    with gr.Row():
        may = gr.Number(label=" MAY")
        jun = gr.Number(label=" JUNE")
        jul = gr.Number(label=" JULY")
        aug = gr.Number(label=" AUGUST")

    with gr.Row():
        sep = gr.Number(label=" SEPTEMBER")
        oct_ = gr.Number(label=" OCTOBER")
        nov = gr.Number(label=" NOVEMBER")
        dec = gr.Number(label=" DECEMBER")

    output = gr.Textbox(label="Prediction Result")
    output_file = gr.File(label="Download Prediction Report (CSV)")
    severity_img = gr.Image(label="District-wise Severity Visualization")

    submit = gr.Button("Predict Flood & Show Severity Map")

    submit.click(
        predict_flood_and_visualize,
        inputs=[year, jan, feb, mar, apr, may, jun, jul, aug, sep, oct_, nov, dec],
        outputs=[output, output_file, severity_img]
    )

demo.launch()
```
